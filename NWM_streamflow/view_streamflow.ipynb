```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
with open("NWM_CHRTOUT_11473900.csv", "r") as f:
    df_11473900 = pd.read_csv(f, index_col=0)
with open("NWM_CHRTOUT_11480390.csv", "r") as f:
    df_11480390 = pd.read_csv(f, index_col=0)
with open("NWM_CHRTOUT_11528700.csv", "r") as f:
    df_11528700 = pd.read_csv(f, index_col=0)
```

```python
df_11473900.head()
```

```python
df_11473900.plot()
df_11480390.plot()
df_11528700.plot()
```

```python
plt_start = '1993-10-01'
plt_end = '1994-09-30'
plt.plot(df_11473900.loc[plt_start:plt_end, 'streamflow'], label='11473900')
plt.plot(df_11480390.loc[plt_start:plt_end, 'streamflow'], label='11480390')
plt.plot(df_11528700.loc[plt_start:plt_end, 'streamflow'], label='11528700')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(4)) 
plt.xticks(rotation=45) 
plt.legend()
plt.show()
plt.close()
```
